```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = sns.load_dataset("tips")
x
```

```python
sns.countplot(x="sex",data=x,hue="smoker")
```

```python
sns.barplot(x="sex",y="total_bill",data=x)                #In bar plot there is always one categorical data(in this case x-axis.) and one quantitative data(in this case y-axis.).
```

The boxplot always takes average (mean). To take other measures such as sum,median etc. we need to specify it using numpy. You can specify using **estimator property.**

---


```python
sns.barplot(x="sex",y="total_bill",data=x,estimator=np.sum)
```

```python
sns.barplot(x="sex",y="tip",data=x,estimator=np.sum)
```

```python
sns.boxplot(x="tip",y="day",data=x,palette='rainbow')
```

```python
sns.violinplot(x="day",y="total_bill",data=x)
```

```python
sns.stripplot(x="day",y="total_bill",data=x,hue="sex",jitter=True)                              #stripplot in simple terms means a scatter plot inside a strip.
```

```python
sns.swarmplot(x="day",y="total_bill",data=x,hue="sex")
```

```python
sns.violinplot(x="tip",y="day",data=x)
sns.swarmplot(x="tip",y="day",data=x)
```
