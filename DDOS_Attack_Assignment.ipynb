## Load the data



```python
import pandas as pd

df = pd.read_csv('/content/Friday-WorkingHours-Morning.pcap_ISCX.csv')
display(df.head())
display(df.shape)
```

## Explore the data


```python
display(df.columns)
```

```python
df.columns = df.columns.str.strip()
df.info()
display(df.describe())
display(df.isnull().sum())
display(df['Label'].value_counts())
```

## Preprocess the data



```python
import numpy as np

# Identify columns with constant values
constant_columns = [col for col in df.columns if df[col].nunique() == 1]
df_cleaned = df.drop(columns=constant_columns)

# Identify and handle infinite values
df_cleaned.replace([np.inf, -np.inf], np.nan, inplace=True)
df_cleaned.dropna(inplace=True)

# Separate features (X) and target (y)
X = df_cleaned.drop(columns=['Label'])
y = df_cleaned['Label']

display(X.head())
display(y.head())
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Encode the categorical target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Scale the numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

display(X_scaled[:5])
display(y_encoded[:5])
```

## Prepare data for modeling



```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

display(X_train.shape)
display(X_test.shape)
display(y_train.shape)
display(y_test.shape)
```

## Choose and train a machine learning model


```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

## Evaluate the model



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```

## Demonstrate ddos detection



```python
# Select a few instances (assuming the first few are BENIGN and some later ones are DDoS based on the value counts)
benign_samples = df_cleaned[df_cleaned['Label'] == 'BENIGN'].head()
ddos_samples = df_cleaned[df_cleaned['Label'] == 'DDoS'].head()
selected_instances = pd.concat([benign_samples, ddos_samples])

# Separate features and labels for selected instances
selected_features = selected_instances.drop(columns=['Label'])
selected_labels = selected_instances['Label']

# Preprocess the selected instances: scale the numerical features using the fitted scaler
selected_features_scaled = scaler.transform(selected_features)

# Predict the labels using the trained model
predicted_labels_encoded = model.predict(selected_features_scaled)
predicted_labels = label_encoder.inverse_transform(predicted_labels_encoded)

# Display the original selected instances and their predicted labels
results_df = selected_instances.copy()
results_df['Predicted_Label'] = predicted_labels

display(results_df[['Label', 'Predicted_Label']])
```
