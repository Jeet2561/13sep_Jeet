```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(['deck' , 'embark_town' , 'alive' , 'class' , 'who' , 'adult_male'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
# Dropping the rows
df.dropna(subset=['embarked'], inplace=True)
```

```python
df.info()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
```

```python
df['sex'] = l.fit_transform(df['sex'])
df['embarked'] = l.fit_transform(df['embarked']) # S=2 , C=0 , Q=1
```

```python
df = df.astype(int)
```

```python
df.head()
```

# **Classification Models**

## **1)  Logistic Regression**

```python
x = df.drop(['survived'] , axis=1)
y = df['survived']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test , y_pred)
```

```python
confusion_matrix(y_test , y_pred)
```

```python
print(classification_report(y_test , y_pred))
```

```python
df
```

## **2) K Nearest Nighbour**

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model.fit(x_train_scaled , y_train)
```

```python
y_knn_pred = knn_model.predict(x_test_scaled)
```

```python
accuracy_score(y_test , y_knn_pred)
```

```python
confusion_matrix(y_test , y_knn_pred)
```

```python
print(classification_report(y_test , y_knn_pred))
```

## **3) Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_model = GaussianNB()
```

```python
nb_model.fit(x_train , y_train)
```

```python
y_nb_pred = nb_model.predict(x_test)
```

```python
y_nb_pred
```

```python
accuracy_score(y_test , y_nb_pred)
```

```python
confusion_matrix(y_test , y_nb_pred)
```

```python
print(classification_report(y_test , y_nb_pred))
```

## **4) Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(random_state=42)
```

```python
dt_model.fit(x_train_scaled, y_train)
```

```python
y_pred_dt = dt_model.predict(x_test_scaled)
```

```python
y_pred_dt
```

```python
accuracy_score(y_test , y_pred_dt)
```

```python
confusion_matrix(y_test , y_pred_dt)
```

```python
print(classification_report(y_test , y_pred_dt))
```

## **5) SVM(Support Vector Machine)**

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(kernel='rbf')
```

```python
model_svm.fit(x_train_scaled , y_train)
```

```python
y_pred_svm = model_svm.predict(x_test_scaled)
```

```python
y_pred_svm
```

```python
accuracy_score(y_test , y_pred_svm)
```

```python
confusion_matrix(y_test , y_pred_svm)
```
