```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=sns.load_dataset('tips')
x
```

```python
df = pd.DataFrame(x)
df
```

## Compare Tipping Behaviour :


---



*   Between smokers vs non-smokers
*   Between male vs female
*   Across Different Days






```python
x['Tip_percentage'] = x['tip'] / x['total_bill'] * 100
x
```

```python
print("Smoker vs Non-Smoker : \n" ,x.groupby('smoker')['Tip_percentage'].mean())
print("\nMale vs Female : \n" ,x.groupby('sex')['Tip_percentage'].mean())
print("\nAcross Different Days : \n" ,x.groupby('day')['Tip_percentage'].mean())
```

```python
sns.boxplot(x='sex',y='Tip_percentage',data=x)
```

```python
sns.barplot(x='smoker',y='Tip_percentage',data=x)
```

```python
sns.violinplot(x='day',y='Tip_percentage',data=x)
```
