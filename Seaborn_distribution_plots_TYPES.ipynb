```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.displot(df['total_bill'] , bins = 20)                   #bins is the size for eg. : in this figure 10 to 20 is one bin.
# Tocheck the distribution of data use HISTOGRAM
```

```python
sns.histplot(df['total_bill'] , bins = 20)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'] , bins = 20, kde=True)                    #kde is kernel density estimator Bell Curve - used to check the skewness.
plt.subplot(1,2,2)
sns.histplot(df['tip'] , bins = 20, kde=True)
```

```python
#To compare two datas together use JOINTPLOT
```

```python
sns.jointplot(x = "total_bill" , y = "tip" , data = df , kind = "scatter")
```

```python
sns.jointplot(x = "total_bill" , y = "tip" , data = df , kind = "kde")
```

```python
sns.jointplot(x = "total_bill" , y = "tip" , data = df , kind = "reg")
```

```python
sns.pairplot(df,hue='sex')                                    # Hue is to give color
#Pairplot analyses all the quantitative data togther
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
sns.rugplot(df['total_bill'])
```

```python
sns.rugplot(df)
```
