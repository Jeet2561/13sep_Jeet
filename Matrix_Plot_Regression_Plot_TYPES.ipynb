```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **HEAT MAP**

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tipscorr = tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),annot=True)
```

# **CLUSTER MAP**

```python
sns.clustermap(tipscorr.corr(),annot=True)
```

# **Pivot Table Heat Map**

```python
pvtfligth = flights.pivot_table(values='passengers',index='month',columns='year')
pvtfligth
```

```python
sns.heatmap(pvtfligth)
```

# **Regression Plot**

```python
x = sns.load_dataset('tips')
sns.lmplot(x='total_bill',y='tip',data=x)
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,hue='sex',palette='coolwarm',markers=['o','v'])
```
