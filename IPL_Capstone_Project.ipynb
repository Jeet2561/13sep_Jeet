```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/IPL.csv")
df
```

```python
df.info()                                                                       #To get info. about the data or to check the data.
```

```python
df.shape
print(f"Columns are {df.shape[0]} and rows are {df.shape[1]}")
```

```python
df.isnull().sum()                           #To see how many null values are there.
```

## **1. Which team won the most matches ?**

```python
match_wins = df['match_winner'].value_counts()
match_wins
sns.barplot(y=match_wins.index , x=match_wins.values)
plt.title("Matches won by each team")
plt.show()
```

### 2. **Toss Decision Trends**

```python
sns.countplot(x = df['toss_decision'])
plt.title("Toss Decision Trends")
plt.show()
```

## 3. **Toss Winner vs Match Winner**

```python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (count * 100) / df.shape[0]
percentage.round(2)
```

## **4. How do teams win ? (Runs vs Wickets)**

```python
sns.countplot(x = df['won_by'])
plt.title("Won By")
plt.show()
```

# **Key Player Performance**

## **1. Most "Player of the Match" Awards**

```python
count = df['player_of_the_match'].value_counts().head(10)
count
sns.barplot(y=count.index , x=count.values)
plt.title("Top 10 Players with Most Player of the Match Awards")
plt.show()
```

## **2. Two Top Scorers**

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high                                                                              #The values are the total of the runs of all the matches and not just one match.
plt.barh(y=high.index , width=high.values)
plt.title("Top Scorers")
plt.show()
```

## **3. 10 Best Bowling Figures**

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind='barh')
plt.title("Top 10 Bowlers")
plt.show()
```

# **Venue Analysis**

## **1. Most Matches Played by Venue**

```python
venue_count = df['venue'].value_counts()
venue_count
sns.barplot(y=venue_count.index , x=venue_count.values,palette='rainbow')
plt.title("Most Matches Played by Venue")
plt.show()
```
